# 树

## 二叉树

```python
# 定义
class TreeNode:
    def __init__(self):
        self.val=val
        self.left :TreeNode|None =None
        self.right:TreeNode|None =None
```

1. 根节点： 二叉树顶层的节点，没有父节点
2. 叶节点： 没有子节点的节点
3. 连接两个节点的指针
4. 层：
5. 度： 子节点数量
6. 高度： 从根节点到最远叶节点所经过边的数量
7. 深度： 从根节点到该节点所经过边的数量
8. 高度： 从距离最远的叶节点到该节点所经过的边的数量。

二叉树类型：
1. 完美二叉树： 所有子节点被充满。
2. 完全二叉树： 除最后一层未填满外，其他都有两个节点，且靠左填充。
3. 满二叉树： 除叶节点外均有两个节点。
4. 平衡二叉树： |左子树高度-右字数高度| <=1

```python
from collections import deque
def level_order(root:TreeNode):
  queue : deque[TreeNode] = deque()
  queue.append(root)
  res=[]
  while len(queue) != 0:
    root=queue.popleft()
    res.append(root.val)
    if root.left is not None:
      queue.append(root.left)
    if root.right is not None:
      queue.append(root.right)
  return res
print(level_order(n1))
```

**前序查找、中序查找、后序查找**
1. 使用迭代思想。
2. 终止条件：root=None 
3. 左子树-> 根节点-> 右子树

```python
res=[]
def pre_order(root:TreeNode):
  if root is None:
    return
  res.append(root.val)
  pre_order(root.left)
  pre_order(root.right)
  return res

def in_order(root:TreeNode,res:list):
  if root is None:
    return
  in_order(root.left,res)
  res.append(root.val)
  in_order(root.right,res)
  return res
def inorder(root:TreeNode):
  res=[]
  return in_order(root,res)

def post_order(root:TreeNode):
  if root is None:
    post_order(root.right)
    res.append(root.val)
    post_order(root.left)
```

查询某个元素
只需要在遍历中添加即可因此放在一起

```python
# 广度搜索 查询
def find(node,val):
  queue : deque[TreeNode]=deque()
  queue.append(node)
  switch=False
  while queue:
    root=queue.popleft()

    if root.val == val:
      switch=True
      break
    if root.left != None:
      queue.append(root.left)
    if root.right != None:
      queue.append(root.right)
  if switch==True:
    print('找着了')
  else:
    print('没有')
#find(n1,9)
# 前序 查找
def find_pre(node,val):
  if node is None:
    return
  if node.val == val:
    print('找到了')
  find_pre(node.left,val)
  find_pre(node.right,val)
#find_pre(n1,4)
```

# 搜索二叉树

2. 所有搜索树 左子树小于右子树
3. 增删查改 ： O(logn)

```python
# 定义 数组-> 搜索二叉树
def init(root:TreeNode,val):
    if root is None:
        return TreeNode(val)
    if root.val < val:
        root.right=init(root.right,val)
    if root.val > val:
        root.left =init(root.left,val)
    return root
def loop_init(arr:list):
    root=None
    for val in arr:
        root = init(root,val)
    return root
```

```python
# 查找 - 广度
def find_search(node:TreeNode,val):
  queue:deque[TreeNode] =deque()
  queue.append(node)
  switch:bool=False
  while queue:
    root =queue.popleft()
    if root.val == val:
      switch=True
      break
    if root.val > val and root.left != None:
      queue.append(root.left)
    if root.val < val and root.right != None:
      queue.append(root.right)
  if switch == True:
    print('找着了')
  else:
    print('没有')
#find_search(Tree,4)
```

```python
# 插入
def insert(node:TreeNode,val):
  if node is None:
    return TreeNode(val)
  if node.val < val:
    node.right=insert(node.right,val)
  else:
    node.left=insert(node.left,val)
  return node
#insert(Tree,10)
#print(inorder(Tree))
```

```python
# 删除
def delete(node:TreeNode,val):
  if node.left is not None:
    if node.left.val == val:
      node.left = node.left.left
      return node
  if node.right is not None:
    if node.right.val == val:
      node.right = node.right.right
      return node
  if node.val < val :
    delete(node.right,val)
  if node.val > val :
    delete(node.left,val)
  return node
#delete(Tree,4)
#print(inorder(Tree))
```
