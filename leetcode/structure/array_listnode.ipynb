# 数组与链表

## 数组

限定：
1. 数组线性储存
2. 连续的内存空间
3. 长度不可变
4. 元素类型相同

```python
# 定义数组
Arr:list[int]=[1]*5
```

访问元素： 元素内存地址= 数组内存地址+元素长度*元素索引
因此元素类型需要相同，因为元素长度需要相同。

```python
def access(Arr:list,pos:int):
    if 0 <= pos <= len(Arr):
        return Arr[pos]
    else:
        raise ValueError("超出边界")
access(Arr,2)
```

插入元素： 如果数组中插入元素，此位置后面元素需要向后移动（会导致元素丢失）

```python
def insert(Arr:list,pos:int,ele:int) -> list:
    if 0 <= pos <= len(Arr):
        for i in range(pos-1,len(Arr)-1):
            Arr[i+1]=Arr[i]
        Arr[pos-1]=ele
        return Arr
    else:
        raise ValueError('超出边界')
insert(Arr,2,3)
```

删除元素： 制定位置后面元素向前移动一位。会有空位留出

```python
def delete(Arr:list,pos:int) -> list:
    if 0 <= pos <= len(Arr):
        for i in range(pos-1,len(Arr)-1):
            Arr[i]=Arr[i+1]
        Arr.pop(pos)
        return Arr
    else:
        raise ValueError('超出边界')
delete(Arr,2)
```

查找元素，线性查找

```python
def find(Arr:list,ele:int)->int:
    switch=False
    for i in Arr:
        if ele == i:
            switch=True
    if switch == True:
        print('找到了')
    else:
        print('没有此元素')

find(Arr,1)
find(Arr,3)
    
```

扩容数组： 初始化一个更大数组，复制进入再返回

```python
def extend(Arr:list,n:int)-> list:
    ans=[0]*n*len(Arr)
    for i in range(0,len(Arr)):
        ans[i]=Arr[i]
    return ans
extend(Arr,2)
```

## 链表

定义
1. 各节点通过‘引用’连接，引用记录了下一个节点的地址
Node对象， 1. 节点的值。2.指向下一个节点的‘引用’。

```python
# 初始化
class ListNode:
    def __init__(self,val:int):
        self.val=val
        self.next : ListNode|None = None
n0=ListNode(1)
n1=ListNode(2)
n2=ListNode(3)
n3=ListNode(4)
n4=ListNode(5)

n0.next=n1
n1.next=n2
n2.next=n3
n3.next=n4
print(n1)
print(n0.next)
print(n0.next==n1)
# 说明 n0 就是指针
```

操作：

```python
# 遍历
def access(node:ListNode,index:int)-> ListNode|None:
  for _ in range(0,index+1):
    node=node.next
  return node
print(access(n0,3).val)

#  查询 
def find(node:ListNode,target:int) :
  while node.next != None:
    if node.val==target:
      return node.next
    node=node.next
print(find(n0,3))
```

```python
# 环结构
class circle:
    def __init__(self,val:int):
        self.val=val
        self.next: ListNode|None =None
n0=circle(0)
n1=circle(1)
n2=circle(2)
n3=circle(3)
n0.next=n1
n1.next=n2
n2.next=n3
n3.next=n0
print(n3.next==n0)

# 双向链表
class BioListNode:
    def __init__(self,val:int):
        self.val=val
        self.pre : BioListNode | None = None
        self.next : BioListNode | None = None
```

## 剑指offer 题

**从尾到头打印链表**

输入一个链表的头节点，按链表从尾到头的顺序返回每个节点的值（用数组返回）。

如输入{1,2,3}的链表

返回一个数组为[3,2,1]

0 <= 链表长度 <= 10000

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def printListFromTailToHead(self , listNode: ListNode) -> list[int]:
        # write code here
        st=[]
        ans=[]
        while True :
            if listNode != None:
                st.append(listNode.val)
            else:
                return ans
            if listNode.next == None:
                break
            else:
                listNode= listNode.next
        for i in range(0,len(st)):
            ans.append(st[len(st)-i-1])

        return ans
            
```

**合并两个排序的链表**

输入两个递增的链表，单个链表的长度为n，合并这两个链表并使新链表中的节点仍然是递增排序的。


```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def Merge(self , pHead1: ListNode, pHead2: ListNode) -> ListNode:
        # write code here

        st=[]
        while pHead1 and pHead2 != None:
            if pHead1.val < pHead2.val:
                st.append(pHead1)
                pHead1=pHead1.next
            else:
                st.append(pHead2)
                pHead2=pHead2.next
        if pHead1 ==None:
            st.append(pHead2)
        else :
            st.append(pHead1)

        for i in range(0,len(st)-1):
            st[i].next =st[i+1]
        return st[0]
```
